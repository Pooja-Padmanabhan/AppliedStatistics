One real-time industry scenario where these concepts of Uniform distribution, Normal Distribution, Binomial Distribution, Poisson Distribution can be used to obtain data driven business soluions is in an insurace company where the evaluation of risk of insuring a large number of people in a particluar region can be taken into account.

### UNIFORM DISTRIBUTION
It can be used to determine the probability of different outcomes when the likelihood of each outcome is the same.example,an insurance company can use uniform distribution to estimate the likelihood of a car accident happning in a particular region by assuming that all drivers have same probability of getting into an accident.

### NORMAL DISTRIBUTION
Normal Distribution can be used to determine the feasibility of an event occuring that follows a bell-shaped curve,such as the frequency of medical claims made by policyholders in a particular region. An insurance company can use normal distribution to analyse the expected number of claims it will receive in a given period.

### BINOMIAL DISTRIBUTION
Binomaial distribution can be used to estimate the probability outcome in a situation where there are only two possible outcomes, such as chances of a policy-holder getting into an accident. The companycan use this type of distribution to evaluate the risk of insuring a particular group of people and rule the probability of certain number of claims being made.

### POISSON DISTRIBUTION
Poisson distribution can be used to asses the probability of a particular event occuring at a particular rate,such as the number of vehicle thefts in a given area. The company can use Poisson distribution to settle the possibility of a particular number of claims being made in a given period.

By analysing data using these statistical methods, an insurance company can figure out the level of risk involved in insuring a particular region, and adjust it's policies and premiums accordingly to mitigate that risk

### A. Is there any significant difference between Funds raised by companies that are still operating vs companies that closed down?

### B. Write the null hypothesis and alternative hypothesis. 

Null hypothesis (Ho) : There is no difference between the two means

Alternate hypothesis (Ha) : There is significant difference between the two means

### C. Test for significance and conclusion 

```python
from statsmodels.stats.weightstats import ztest

sample1 = data.loc[data.OperatingState == 'Operating', 'Funds_in_million']
sample2 = data.loc[data.OperatingState =='Closed', 'Funds_in_million']

alpha = 0.05   # Let's consider a significance level of 5%

test_statistic, p_value = ztest(sample1, sample2)

if p_value <= alpha:
    print(f'Since the p-value, {round(p_value, 3)} < {alpha} (alpha) the difference is significant and,Conclusion we reject the Null hypothesis')
else:
    print(f'''\t Since the p-value, {round(p_value,3)} > {alpha} (alpha) the difference is not significant and,
    \t Conclusion: we fail to reject the Null hypothesis''')
    
```
